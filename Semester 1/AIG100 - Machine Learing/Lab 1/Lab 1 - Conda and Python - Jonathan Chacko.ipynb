
# Basic File Operations in Python

File operations are fundamental for data processing. In this section, we'll cover how to read from and write to files in Python.

## Task
Write a Python script to read a text file and count the frequency of each word in the file.

## Assessment Question
- Discuss the different modes of opening a file in Python.


```python
## Task


import re
with open(r"C:\Users\JonathanChackoPattas\OneDrive - Maritime Support Solutions\Desktop\Class Notes\Seneca\AIG100 - Machine Learing\Lab 1\corpus5.txt", 'r') as file: # , encoding='utf-8'
    file_contents = file.read()
file_contents = (lambda s: re.sub(r'([^a-zA-Z0-9\s])', r' \1 ', s).strip())(file_contents)
file_contents = file_contents.split()
file_contents = list(map(lambda item: ''.join(char for char in item if char.isalnum() or char.isspace()), file_contents))
file_contents = [word.lower() for word in  file_contents if word != ""]
word_count = {word: file_contents.count(word) for word in set(file_contents)}
word_count
```

```python
## Assessment


"""
Depending on the Task we need to use the File for there are Differnt Modes to Open a File.
r - Read Mode
w - Write Mode (Creates File if Doesn't Exist)
a - Append Mode
x - Exclusive Creation Mode (fails if file already exists)
r+ - Read and Write Mode
a+ - Read and Append Mode
w+ - Write and Read Mode
b - Binary Mode
t - Text Mode (default)
"""
```
