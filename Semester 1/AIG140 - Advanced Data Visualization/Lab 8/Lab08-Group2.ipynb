```python
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. Importing Data and Matplotlib

```python
df = pd.read_csv("company_sales_data.csv")
df
```

### 2. Creating Line Plots

```python
x = df['month_number']
y = df['total_profit']

plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.xlabel("Month Number") 
plt.ylabel("Total Profit") 
plt.title("Monthly Total Profit") 
plt.grid(True)
plt.show()
```

### 3. Multiple Line Plots

```python
y1 = df['facewash']
y2 = df['bathingsoap']

plt.plot(x, y1, marker='o', color='sienna', label='Face wash')
plt.plot(x, y2, marker='^', label='Bathing soap')
plt.title("Face Wash vs. Bathing Soap Monthly Sales") 
plt.legend()
plt.grid(True)
plt.show()
```

### 4. Bar Plots

```python
plt.bar(x, y, color='darkorchid')
plt.xlabel("Month Number"), plt.ylabel("Total Profit") 
plt.title("Total Profit per Month") 
plt.show()
```

### 5. Scatter Plots

```python
x = df['facecream']
y = df['total_profit']

plt.scatter(x, y, color='lightseagreen', marker='o')
plt.xlabel("Face Cream Units")
plt.ylabel("Total Profit")
plt.title("Face Cream vs Total Profit")
plt.grid(True)
plt.show()
```

### 6. Pie Chart

```python
labels = [clmn for clmn in df.columns if clmn not in ['month_number', 'total_units', 'total_profit']]
values = [df[clmn].sum() for clmn in labels]

plt.figure(figsize=(10, 10))
plt.pie(values, labels=labels, autopct='%1.1f%%')
plt.title("Percentage of Total Sales by Product")
plt.legend()
plt.show()
```

### 7. Histogram

```python
data = df['total_profit']

plt.figure(figsize=(10, 10))
plt.hist(data, bins=6, color='lightpink', edgecolor='black')
plt.xlabel("Total Profit ($)")
plt.ylabel("Frequency")
plt.title("Distribution of Monthly Total Profits")
plt.show()
```
