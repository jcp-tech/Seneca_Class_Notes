10. Write code to find how many times does the word ‘king’ (all lower case) appear in text6? Write code to calculate, how much is this as a percentage of the total number of words in this text? (Do so using only tools appearing in the previous questions, and elementary Python tools; do not use FreqDist yet!)

```python
total_words_text6 = len(text6)
```

```python
king_count_lower = text6.count('king')
percentage_king_lower = (king_count_lower / total_words_text6) * 100
print(f"Count of 'king': {king_count_lower}")
print(f"Percentage of 'king': {percentage_king_lower:.2f}%")
print("Note: These Results are case-sensitive.")
```

15. From text 6, display all words that are longer than seven characters, that occur more than seven times. (The frequency distribution you created for this text is useful here!)

```python
fdist3 = FreqDist(text6)
long_frequent_words = [word for word, count in fdist3.items() if len(word) > 7 and count > 7]
sorted(long_frequent_words)
```
