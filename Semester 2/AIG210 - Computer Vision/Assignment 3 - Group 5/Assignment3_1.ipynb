```python
import cv2, os
import numpy as np
from matplotlib import pyplot as plt
script_dir = os.getcwd() # Get the directory of the current script
```

```python
def ipynb_cv_imshow(title, img, cv=True):
    """
    Display an image in a Jupyter Notebook cell.
    """
    if img is None:
        print("Image not found or could not be read.")
        return
    plt.figure(figsize=(10, 10))
    plt.title(title)
    if cv:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    else:
        plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
credit_image = cv2.imread(os.path.join(script_dir, 'Credit.PNG'))
ipynb_cv_imshow("Original Image", credit_image)
```

```python
gray_image = cv2.cvtColor(credit_image, cv2.COLOR_BGR2GRAY)
ipynb_cv_imshow("Gray Image", gray_image)
```

```python
sobel_8u_10 = cv2.Sobel(gray_image, cv2.CV_8U, 1, 0, ksize=3)
sobel_8u_01 = cv2.Sobel(gray_image, cv2.CV_8U, 0, 1, ksize=3)
sobel_8u_11 = cv2.Sobel(gray_image, cv2.CV_8U, 1, 1, ksize=3)
ipynb_cv_imshow("Sobel dx = 1 & dy = 0 (8U)", sobel_8u_10)
ipynb_cv_imshow("Sobel dx = 0 & dy = 1 (8U)", sobel_8u_01)
ipynb_cv_imshow("Sobel dx = 1 & dy = 1 (8U)", sobel_8u_11)
```

```python
sobel_64f_10 = cv2.Sobel(gray_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_64f_10_abs = cv2.convertScaleAbs(sobel_64f_10)
sobel_64f_01 = cv2.Sobel(gray_image, cv2.CV_64F, 0, 1, ksize=3)
sobel_64f_01_abs = cv2.convertScaleAbs(sobel_64f_01)
sobel_64f_11 = cv2.Sobel(gray_image, cv2.CV_64F, 1, 1, ksize=3)
sobel_64f_11_abs = cv2.convertScaleAbs(sobel_64f_11)
ipynb_cv_imshow("Sobel dx = 1 & dy = 0 (64F)", sobel_64f_10_abs)
ipynb_cv_imshow("Sobel dx = 0 & dy = 1 (64F)", sobel_64f_01_abs)
ipynb_cv_imshow("Sobel dx = 1 & dy = 1 (64F)", sobel_64f_11_abs)
```

```python
# Apply Canny edge detector with two different sets of thresholds
canny_low1, canny_high1 = 50, 150
canny_low2, canny_high2 = 100, 200
```

```python
canny_edges_1 = cv2.Canny(gray_image, canny_low1, canny_high1)
ipynb_cv_imshow("Canny Edges (Low: {}, High: {})".format(canny_low1, canny_high1), canny_edges_1)
```

```python
canny_edges_2 = cv2.Canny(gray_image, canny_low2, canny_high2)
ipynb_cv_imshow("Canny Edges (Low: {}, High: {})".format(canny_low2, canny_high2), canny_edges_2)
```
